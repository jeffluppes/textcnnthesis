### Ontology -> Wikipedia data scraper

```python
#imports
import wikipedia
import requests
import random
import pickle
from lxml import etree
from lxml.etree import fromstring

#configs
endpoint = ''
output_folder = '../productontology/'

titles = []

#get the OWL file 
for i in range(24):
    # the dump API takes two parameters, x (how many items) and y (offset) thus we need 
    x = 1000
    y = i*x 
    query =  str("http://www.productontology.org/dump.rdf?limit="+str(x)+"&offset="+str(y)+"")
    print(query)
    req = requests.request('GET', query)
    
    # get the payload
    a = req.content
    # get the xml (rdf) tree
    b = etree.fromstring(a)

    temptitles = ""
    for c in b:
        if(c.tag =='{http://www.w3.org/2002/07/owl#}Class'): 
            # split on / and reverse the string, so that the last element (the title) is now the first element
            # this is useful because we dont know the length of the string in advance!
            f = str(c.attrib).split('/')[::-1]
            # at this point the title still includes '} at the end, so we do not request the last two items
            temptitles += f[0][:-2]+ " "
    
    
    #split on whitespace
    temptitles = temptitles.split(" ")
    #remove the first two and the last element (these are always the result of markup artifacts)
    #print("removing entry 1002:"+ str(temptitles[-1]))
    del temptitles[-2]
    #print("removing entry 1:"+ str(temptitles[1]))
    del temptitles[1]
    #print("removing entry 0:"+ str(temptitles[0]))
    del temptitles[0]
    print(len(set(temptitles)))
    #finally, add the temptitles to our titles list
    titles = list(set(titles + temptitles))
    print("added "+str(len(temptitles))+" to our list! We now have "+str(len(titles))+" titles of products.")    
```

```python
#store the resulting list to file
with open('pickle_product_names_23k', 'wb') as fp:
    pickle.dump(titles, fp)
```

```python
with open('productnamesdump.txt', 'w') as f:
    for item in titles:
        f.write("%s\n" % item)
```

```python
#show some random titles
randoms = random.sample(range(len(titles)), 10)
titles = list(titles)
for i in randoms:
    print(titles[i])
```

This is not perfect, e.g. `New_Britain,_Connecticut` exists as a product or service, apparently. And we have entries for `Dubliner_Cheese` as well as `Dubliner_cheese`... Yet, most of these are indeed product descriptions which make this a lot better than the overal dump of wikipedia!

```python
corpus = ""
fails = 0
for idx, title in enumerate(titles):
    if title != "":
        if (idx % 10 == 0):
            print(str(idx)+"/"+str(len(titles)))
            
        try:
            corpus = wikipedia.page(str(title)).summary
            #save and store
            name = str(output_folder + title+'.txt')
            with open(name, 'w') as f:
                f.write(corpus)
        except:
            fails += 1
            
print(fails) #5325.
```
